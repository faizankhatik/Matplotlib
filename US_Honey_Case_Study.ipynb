# 1) Which states are rarely contributing to honey production for the last 27 years?

```python
# 27 years from 1995 to 2021
# Which state have not contributed in honey for all the 27 years
# If any state is contributing in honey production that means it data will be present in year column
# Lets say alaba contributed in 1995 , 1980 , 2000 that means alabab data will be present 3 times in our dataset
# So you need to find the count of states like alababa data is present how many time , california data is present how many time

df["state"].value_counts()
```

# 2) Which are the top 5 Honey producing states in the US ?

```python
# Which states are having the highest honey production in all 27 years
# Means you need to add the honey production for all states for all 27 yeasr ten compare the final honey production.

plt.figure(figsize=(20,10))
sns.barplot(x=df["state"] , y=df["production"])
plt.xlabel("State names")
plt.ylabel("Total production of honey")
plt.title("Honey production of all states")
plt.xticks(rotation=90)  # rotate the value on x axis by 90degree
plt.show()
```

**Top 5 honey producing state**

1) North dakota

2) califronia

3) south dakota

4) Florida

5) Montata

```python
# I need to compare my state name with total production of that state in 27 years
# then i can sort the data in descending order so the state with highest honey production will come at top
new_df=df.groupby("state").sum(["production"]).sort_values("production", ascending = False).reset_index()
new_df
```

# 3) What is the Change in mean Average price of Honey from 1995 to 2021?

```python
# We need to find the difference of average price of honey of all state in 1995 and 2021
new_df = df.groupby("year").mean(["average_price"]).reset_index()
new_df
```

```python
new_df[["year","average_price"]]

# in 1995 average price of all state honey production was 74.840909
# in 2021 average price of all state honey production was 3.334250
```
